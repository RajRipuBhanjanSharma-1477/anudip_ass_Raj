### 1. Suppose you have a dataset containing daily temperature readings for a city, and you want to identify days with extreme temperature conditions. Find days where the temperature either exceeded 35 degrees Celsius (hot day) or dropped below 5 degrees Celsius (cold day). Input: 

temperatures = np.array([32.5, 34.2, 36.8, 29.3, 31.0, 38.7, 23.1, 18.5, 22.8, 37.2]) 

```python
temperatures = np.array([32.5, 34.2, 36.8, 29.3, 31.0, 38.7, 23.1, 18.5, 22.8, 37.2])
```

```python
import numpy as np
import pandas as pd
days=np.array(range(1,len(temperatures)+1))
type=[""]*len(temperatures)
def wether(x)->str:
    if x>35:
        return "hot"
    elif x < 5:
        return "cold"
    else :
        return "normal"
for x in range(len(temperatures)):
    type[x]=wether(temperatures[x])
data={"days":days,"temperatures":temperatures,"Wether":type}
data=pd.DataFrame(data)
```

```python
data
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def color(x)->str:
    if x>35:
        return "red"
    elif x < 5:
        return "blue"
    else :
        return "yellow"
# Plot
plt.figure(figsize=(8,5))
plt.bar(data["days"],data["temperatures"], color=[color(x) for x in data["temperatures"] ])

# Labels and title
plt.xlabel("Days")
plt.ylabel("temperature")
plt.title("Weather Trend Over 10 Days")
plt.xticks(data["days"])  # Show all days on X-axis
# plt.yticks(wether(data["temperatures"]))  # Show categorical labels for Y-axis
for x in range(len(data["temperatures"])):
    plt.annotate(wether(data["temperatures"][x]),xy=(data["days"][x],data["temperatures"][x]),rotation=45,va='bottom', ha='center')

# Show plot
plt.show()
```

### 2. Suppose you have a dataset containing monthly sales data for a company, and you want to split this data into quarterly reports for analysis and reporting purposes. 

Input: 

monthly_sales = np.array([120, 135, 148, 165, 180, 155, 168, 190, 205, 198, 210, 225]) 

```python
monthly_sales=np.array([120, 135, 148, 165, 180, 155, 168, 190, 205, 198, 210, 225])
```

```python
month=np.array(range(1,len(monthly_sales)+1))
```

```python
quaterly_month_sales=monthly_sales.reshape(4,3)
quaterly_month=month.reshape(4,3)
data={}
for i in range(4):
    data["Quater_"+str(i)]={"months":quaterly_month[i],"Monthly_Sales":quaterly_month_sales[i]}
data=pd.DataFrame(data)
data
```

```python
import matplotlib.pyplot as plt
col=["red","yellow","green","blue"]
for x in range(4):
    for y in range(3):
        plt.bar(data["Quater_"+str(x)][0][y],data["Quater_"+str(x)][1][y],color=col[x])
        plt.annotate("Q"+str(x),xy=(data["Quater_"+str(x)][0][y],data["Quater_"+str(x)][1][y]),rotation=45,va="bottom",ha="center")
plt.xlabel("Quater")
plt.ylabel("Sales")
plt.title("Quaterly sales")
plt.show()
```

### 3. suppose you have a dataset containing customer data, and you want to split this data into two groups: one group for customers who made a purchase in the last 30 days and another group for customers who haven't made a purchase in the last 30 days. 

Input: 

customer_ids = np.array([101, 102, 103, 104, 105, 106, 107, 108, 109, 110]) 

last_purchase_days_ago = np.array([5, 15, 20, 25, 30, 35, 40, 45, 50, 55]) 

```python
customer_ids = np.array([101, 102, 103, 104, 105, 106, 107, 108, 109, 110])
last_purchase_days_ago = np.array([5, 15, 20, 25, 30, 35, 40, 45, 50, 55]) 
```

```python
len(customer_ids),len(last_purchase_days_ago)
```

```python
import pandas as pd
a=[]
b=[]
for x in range(len(last_purchase_days_ago)) :
    if last_purchase_days_ago[x]>30:
        a.append(customer_ids[x])
    else:
        b.append(customer_ids[x])
data={
    "made_a_purchase_in_last_30_days":b,
    "did'nt_made_a_purchase_in_last_30_days":a
}
data=pd.DataFrame(data) 
data   
```

### 4. Suppose you have two sets of employee data one containing information about full-time employees and another containing information about part-time employees. You want to combine this data to create a comprehensive employee dataset for HR analysis. Input: 

# Employee data for full-time employees 

full_time_employees = np.array([ [101, 'John Doe', 'Full-Time', 55000], [102, 'Jane Smith', 'Full-Time', 60000], [103, 'Mike Johnson', 'Full-Time', 52000] ]) 

# Employee data for part-time employees 

part_time_employees = np.array([ [201, 'Alice Brown', 'Part-Time', 25000], [202, 'Bob Wilson', 'Part-Time', 28000], [203, 'Emily Davis', 'Part-Time', 22000] ])

```python
full_time_employees = np.array([ [101, 'John Doe', 'Full-Time', 55000], [102, 'Jane Smith', 'Full-Time', 60000], [103, 'Mike Johnson', 'Full-Time', 52000] ]) 
part_time_employees = np.array([ [201, 'Alice Brown', 'Part-Time', 25000], [202, 'Bob Wilson', 'Part-Time', 28000], [203, 'Emily Davis', 'Part-Time', 22000] ]) 
```

```python
import pandas as pd
col=["Employees_id","Employees_Name","Employees_type","Employees_Salary"]
data={}
for i in range(len(col)):
    tem=[]
    for x in range(len(full_time_employees)):
        tem.append(full_time_employees[x][i])
    for y in range(len(part_time_employees)):
        tem.append(part_time_employees[y][i])
    data[col[i]]=tem
data=pd.DataFrame(data)
data    
```
